Creare un notebook jupyter che a partire dal testo contenuto nel file ___"J_Verne_girodelmondoin80giorni.txt"___:

1. crei un dizionario delle frequenze per ciascun carattere (**line feed escluso**); 
2. crei il dizionario delle adiacenze per ciascun carattere (**line feed escluso**);
3. **[FACOLTATIVO]** crei un indice dei capitoli. Ogni capitolo inizia riportando il numero romano del capitolo e il suo titolo in questo modo:

>CAPITOLO XXIX.
>
>Nel quale si narrano cose che capitano unicamente sulle ferrovie
>americane.

```python
f = open("J_Verne_girodelmondoin80giorni.txt", 'r')
testo = f.readlines()
f.close()
dizionario_adiacenze = {}
dizionario_caratteri = {}
alfabeto = "abcdefghijklmnopqrstuvwxyz"
```

```python
punteggiatura = ["!", "€", "™", "¹" "¬", "=", "¨", "'", ".", "?", "-", "_", "(", ")", ";", "«", "»", "[", "]", "{", "}", ":", "\"", ",", "^"]
for i in range(len(testo)):
    for p in punteggiatura:     #se c'è un carattere speciale devo toglierlo e mettere uno spazio al posto
        testo[i] = testo[i].replace(p, " ")
    testo[i] = testo[i][:-1]    #non considero il \n e unisco per evitare di avere troppi spazi
```

```python
testo
```

```python
dizionario_caratteri = {}
for i in range(len(testo)):
    riga = len(testo[i])
    for j in range(riga):
        if testo[i][j].lower() in alfabeto or testo[i][j].lower() == " ":
            if testo[i][j].lower() not in dizionario_caratteri:
                dizionario_caratteri[testo[i][j].lower()] = 1
            else:
                dizionario_caratteri[testo[i][j].lower()] += 1  
```

```python
lista_lettere = []
for riga in testo:
    riga = riga.split()
    #print(riga)
    for lettera in riga:
        for carattere in lettera:
            lista_lettere.append(carattere.lower())
lista_lettere
```

```python
dizionario_adiacenze = {}
for i, carattere in (enumerate(lista_lettere[:-1])):
    if carattere in dizionario_adiacenze and (carattere in alfabeto or carattere == " "):
        if lista_lettere[i+1] not in dizionario_adiacenze[carattere] :
            dizionario_adiacenze[carattere].append(lista_lettere[i+1])
    else:
        dizionario_adiacenze[carattere] = [lista_lettere[i+1]]
dizionario_adiacenze
```
