```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
secondi = 2*60*60
```

```python
x_y = np.random.choice(["x+1", "x-1", "y+1", "y-1", ], secondi)
x_y
```

```python
x = [0]
y = [0]
posizione = (0,0)
for i in range(x_y.size):
    if x_y[i] == "x+1":
        x.append(x[-1] + 1)
        y.append(y[-1])
    elif x_y[i] == "x-1":
        x.append(x[-1] - 1)
        y.append(y[-1])
        
    elif x_y[i] == "y+1":
        y.append(y[-1] + 1)
        x.append(x[-1])
        
    elif x_y[i] == "y-1":
        y.append(y[-1] - 1)
        x.append(x[-1])
```

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.scatter(x, y, marker = 'v', c = "green", s=5)
#ax.plot(x,y, color = 'grey', linewidth=0.2)
plt.show()
```
