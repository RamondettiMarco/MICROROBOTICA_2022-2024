openData = https://data.worldbank.org/indicator/SP.POP.TOTL?end=2021&start=1960

```python
import matplotlib.pylab as plt
import numpy as np
```

```python
#creo array con gli anni da 1960 a 2021 essendo presenti sul file
anni = np.arange(1960,2022)
```

```python
f = open("API_SP.POP.TOTL_DS2_en_csv_v2_4701113/API_SP.POP.TOTL_DS2_en_csv_v2_4701113.csv", "r")
testo = f.readlines()

lista = [testo[4].replace('"',"").strip().split(",")]
#tralascio le prima 5 righe in quanto non servono.
testo = testo[5:]
#testo
```

```python
paesi = []
for i in range(len(testo)):
        paesi.append(testo[i].replace('"', "").strip().split(","))
        
for i in range(len(paesi)):
    for k in range(len(paesi[i])):
        #print(paesi[i][k])
        if(str(paesi[i][k]) == "SP.POP.TOTL"):
            paesi[i] = paesi[i][k+1:]
            break
```

```python
#calcolo popolazione di un solo anno
"""popolazione = 0
for i in range(len(paesi)):
    if(paesi[i][2] != ''):
        popolazione = popolazione + int(paesi[i][2])
popolazione"""
```

```python
popolazione_annuale = []
popolazioneAnno = 0
for i in range(len(paesi[i])-1):
    popolazioneAnno = 0
    for k in range(len(paesi)):
        if paesi[k][i] != '':
            popolazioneAnno = popolazioneAnno + int(paesi[k][i])
    popolazione_annuale.append(popolazioneAnno)
#popolazione_annuale
```

```python
x = np.linspace(anni[11], anni[-1], len(anni))
f1 = popolazione_annuale
```

```python
fig,ax = plt.subplots(figsize = (15,10))
ax.set_title(f"Popolazione annuale mondiale da {anni[11]} - {anni[-1]}")
ax.plot(x, f1, label="Popolazione annuale")
ax.set_ylabel("POPOLAZIONE, miliardi")
ax.set_xlabel("ANNO")
ax.legend()
plt.show()
```
