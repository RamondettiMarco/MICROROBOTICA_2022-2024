### Cielo stellato

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_star = 1000
```

```python
y = np.random.uniform(0., 1., n_star)
x = np.random.uniform(0., 1., n_star)
y = np.random.choice(np.arange(0.,1.), n_star, replace = False)
```

```python
colori = np.random.uniform(0., 1., (n_star, 3))
trasparenza = np.random.uniform(0., 1., n_star)
dimensione = np.random.uniform(10., 200., n_star)
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.set_facecolor("#123a5b")
ax.scatter(x,y, marker = "*", c="white", s = dimensione, alpha = trasparenza)
ax.set_title("Cielo stellato")
plt.xticks([])
plt.yticks([])
plt.show()
```
