```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('penguins_size.csv', 'r')
file = f.readlines()
f.close()
intestazione = file[0]
intestazione = intestazione.split(',')
intestazione[-1] = intestazione[-1].strip()
print(intestazione)
file = file[1:-1]
```

```python
dizionario = {}    #con chiave intestazione e lista come valori
for chiave in intestazione:
    dizionario[chiave] = []
    
dizionario
```

```python
array_pinguino = []
for riga in file:
    pinguino = riga.split(",")
    pinguino[-1] = pinguino[-1].strip()
    array_pinguino.append(pinguino)
    if "NA" not in pinguino and "." not in pinguino:
        for i, chiave in enumerate(intestazione):
            if i == 0 or i == 1 or i == 6:
                dizionario[chiave].append(pinguino[i])
            else:
                dizionario[chiave].append(float(pinguino[i]))
array_pinguino = np.array(array_pinguino)
```

```python
for chiave in intestazione:
    dizionario[chiave] = np.array(dizionario[chiave])

dizionario
```

### ISTOGRAMMA

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.hist(dizionario[intestazione[-1]], color="orange")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.hist(dizionario[intestazione[0]], color="blue")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.hist(dizionario[intestazione[5]], color="pink")
ax.set_xlabel(intestazione[5])
plt.show()
```

### Scattern

```python
fig, ax = plt.subplots(figsize = (32,16))
ax.scatter(dizionario[intestazione[2]], dizionario[intestazione[3]], c = "red")
ax.set_xlabel(intestazione[2], fontsize = 25)
ax.set_ylabel(intestazione[3], fontsize = 25)
plt.show()
```

#### Lunghezza pinna rispetto alla lunghezza corporea

```python
fig, ax = plt.subplots(figsize = (32, 16))
ax.scatter(dizionario[intestazione[5]], dizionario[intestazione[4]], c = "green")
ax.set_xlabel(intestazione[5], fontsize = 25)
ax.set_ylabel(intestazione[4], fontsize = 25)

plt.show()
```

### SOTTOGRAFICI

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize = (16,8))
axs[0].set_title(f"Grafico: {intestazione[5]}")
axs[0].hist(dizionario[intestazione[5]], color ="yellow") #["pink", "green", "blue", "orange", "yellow"])
axs[0].set_xlabel(intestazione[5])


axs[1].set_title(f"Grafico: {intestazione[5]}, {intestazione[4]}")
axs[1].scatter(dizionario[intestazione[5]], dizionario[intestazione[4]], c = "green", s = 10)
axs[1].set_xlabel(intestazione[5])
axs[1].set_ylabel(intestazione[4])
plt.show()
```
