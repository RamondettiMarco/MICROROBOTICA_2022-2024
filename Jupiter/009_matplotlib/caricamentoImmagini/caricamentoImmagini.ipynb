### Caricamento e visualizzazione immagini presenti nella cartella

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_immagini = 4
fig, ax = plt.subplots(4, 1, figsize = (16,8))
for i in range(n_immagini):
    immagine = plt.imread(f"img{i}.JPG")
    ax[i].imshow(immagine)
plt.show()
```
