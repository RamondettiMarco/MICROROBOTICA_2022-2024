```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1., 1., 101)    #se metto 101 becca lo 0 --> da errore ma elimina il punto calcolato su 0 e funziona lo stess0
f1 = 1/x
```

```python
fig, ax = plt.subplots(figsize = (6, 6))    #dimensione in pollici
ax.grid("both")    #devo dire su quali assi, both entrambi
ax.plot(x, f1, label="Funzione f1: 1/x.")
ax.legend()
plt.show()
```

### Matplotlib

```python
n_star = 20
x = np.random.uniform(0., 1., n_star)
y = np.random.uniform(0., 1., n_star)
```

```python
fix, ax = plt.subplots(figsize = (16, 8))
ax.set_facecolor("black")
#ax.plot(x, ...)  #--> se non passo array ascisse me lo crea lui automaticamente.
ax.plot(x, y, marker = "*", markersize = 20, linestyle = '', color = "white")
plt.xticks([]) #per togliere i valori sull'asse x
plt.yticks([]) #per togliere i valori sull'asse y
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)
s1 = np.sin(x)
s2 = np.cos(x)
s3 = x*0
```

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.set_title("Grafico sin(x) e cos(x)", fontsize = 20)
ax.plot(x, s1, linestyle = ":", linewidth = 3, color = "orange", label = "sin(x)")
ax.plot(x, s2, linestyle = "-.", linewidth = 3, color = "green", label = "cos(x)")
ax.plot(x, s3, linestyle = "--", linewidth = 3, color = "blue", label = "x=x*0")

ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.tick_params(axis = "both", labelsize = 20)
ax.legend(fontsize = 20)  #loc = n --> posso decidere dove mettere la mia legenda, se non setto il parametro me lo mette in automatico nel posto più libero.
ax.grid()
plt.savefig("./graficoSinCos.png", dpi = 200)    #se salva male aumento dimensione figsize.
plt.show()
```

### GRAFICI A DISPERSIONE: SCATTER 

```python
n_star = 500
x = np.random.uniform(0., 1., n_star)
y = np.random.uniform(0., 1., n_star)
colori = np.random.uniform(0., 1., (n_star, 3))
#colore ha un livello di trasparenza Alpha --> 0 trasparente 1 opaco, le vie di messo rendono trasparente il colore. 
trasparenza = np.random.uniform(0., 1., n_star)
dimensione = np.random.uniform(100., 1000., n_star)
```

```python
fig, ax = plt.subplots(figsize = (16,8))
#ax.set_facecolor("black")
ax.scatter(x, y, c = colori, s = dimensione, alpha = trasparenza, marker = 'o')  # c = colori per colorare i punti di colori diversi. 
                                                           #alpha = a un valore di trasparenza, posso usare sia un array di colori che un singolo colore
plt.show()
```
