#### Esercizio 1
Realizza un grafico che rappresenti le funzioni x^2, x^3, x^4, ciascuna con un colore diverso.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 200)
```

```python
F1 = x**2
F2 = x**3
F3 = x**4
```

```python
fig, ax = plt.subplots(figsize = (10,5))    #dimensione in pollici
ax.set_title("Rappresentazione x^2, x^3, x^4")
ax.plot(x, F1, label = "funzione F1: x**2")
ax.plot(x, F2, label = "funzione F2: x**3")
ax.plot(x, F3, label = "funzione F3: x**4")
ax.set_xlabel("ASCISSA")
ax.set_ylabel("ORDINATA")
ax.legend()
plt.show()
```

#### Esercizio 2
Realizza un grafico che rappresenti le rette con coefficienti angolari della lista python: [0, 1/2, 2, 3, 4]. A ogni retta assegna uno stile di linea diverso

```python
lista = [0, 1/2, 1, 2, 4]
x = np.linspace(-10, 10, 10)
```

```python
F0 = lista[0]*x
F1 = lista[1]*x
F2 = lista[2]*x
F3 = lista[3]*x
F4 = lista[4]*x
```

```python
_, ax = plt.subplots(figsize = (10,5))
plt.title("Rappresentazione rette con coefficienti angolari lista python")
ax.plot(x, F0, label= "funzione F0", linestyle = '-')
ax.plot(x, F1, label= "funzione F1", linestyle = '--')
ax.plot(x, F2, label= "funzione F2", linestyle = '-.')
ax.plot(x, F3, label= "funzione F3", linestyle = ':')
ax.plot(x, F4, label= "funzione F4", marker= '1', linestyle = '')
ax.legend()
plt.show()
```
