```python
import numpy as np
```

```python
campioni = 1000
error = 9999.9
array = np.random.normal(19,5,campioni)
array
```

```python
array_error = np.random.randint(0, 1000, size=20)     #potrei avere dei doppioni perchè ciò che estrae lo rimette "dentro"
array_error
```

```python
#uso il choice per non avere possibili doppioni nella random delle 20 posizioni
indici_error = np.random.choice(np.arange(0,1000), size = 20, replace = False)        #replace toglie l'elemento che trovo
indici_error
```

```python
array[indici_error] = error
```

```python
array[indici_error]
```

```python
dati = []
for i in range(0, campioni, 5):
    dati = array[i:i+5] == error
    if np.count_nonzero(dati) > 1:
        print(f"Ci sono almeno {np.count_nonzero(dati)} misurazioni errate in una sequenza di 5 misurazioni. LA CENTRALINA DEVE ESSERE RIPARATA.\nPosizioni: {i}:{i+5}")
```

#### Pulizia codice da 9999.9

```python
array_ripulito = array[array != error]
print(f"Array ripulito: {array_ripulito}")
```

#### Calcolo media

```python
media = np.mean(array_ripulito)
print(f"Media valori: {media}")
```

#### Calcolo minimo

```python
minimo = np.min(array_ripulito)
print(f"Valore minimo: {minimo}.")
```

#### Calcolo massimo

```python
massimo = np.max(array_ripulito)
print(f"Valore massimo: {massimo}.")
```
