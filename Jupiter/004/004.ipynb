### Esercizio

```python
f = open('sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt', 'r')
righe = f.readlines()
f.close()
```

```python
#righe
```

```python
punteggiatura = ["!", "'", ".", "?", "-", "_", "(", ")", ";", "«", "»", "[", "]", "{", "}", ":", "\"", ",", "^"]
```

```python
for i in range(len(righe)):
    for p in punteggiatura:     #se c'è un carattere speciale devo toglierlo e mettere uno spazio al posto
        righe[i] = righe[i].replace(p, " ")
    righe[i] = righe[i][:-1].strip()    #non considero il \n e unisco per evitare di avere troppi spazi
```

```python
dizionario = {}
```

```python
for i in range(len(righe)):
    riga = righe[i].lower().split()
    print(riga)
    for s in range(len(riga)):
        if riga[s] != '':     #se la posizione i della lista è vuoto non lo considero
            if riga[s] not in dizionario:       
                dizionario[riga[s]] = 1      #se la parola non c'è la inserisco e imposto a 1
            else:    
                dizionario[riga[s]] +=1      #se la parola c'è aggiorno il cnt 
```

```python
dizionario
```

```python
#2. dizionario in cui la chiave sono tutte le parole, il valore è una lista che contiene tutte le parole successive
```

```python
dizionario_successivo = {}
```

```python
lista_parole = []
```

```python
for riga in righe:
    riga = riga.split()
    for parola in riga:
        lista_parole.append(parola.lower())
```

```python
#lista_parole
```

for chiave in dizionario:
    dizionario_successivo[chiave] = lista_parole[1]

```python
for i, parola in (enumerate(lista_parole[:-1])):
    #print(i,parola)
    if parola in dizionario_successivo:
        if lista_parole[i+1] not in dizionario_successivo[parola]:
            dizionario_successivo[parola].append(lista_parole[i+1])
    else:
        dizionario_successivo[parola] = [lista_parole[i+1]]
```
